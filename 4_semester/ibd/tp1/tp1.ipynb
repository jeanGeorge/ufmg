```python
# 1 - ok
df = pd.read_sql_query("SELECT * FROM FUNCAO WHERE COD_FUNCAO IN (SELECT COD_FUNCAO FROM DETALHAMENTO_FUNCAO WHERE COD_SUBFUNCAO = (SELECT COD_SUBFUNCAO FROM DETALHAMENTO_FUNCAO GROUP BY COD_SUBFUNCAO ORDER BY COUNT(*) DESC LIMIT 1)) ORDER BY COD_FUNCAO",conn)
df
```

```python
# 5 - ok
df = pd.read_sql_query("SELECT COD_ORGAO, NOME_ORGAO FROM ORGAO NATURAL JOIN (SELECT COD_ORGAO, SUM(VALOR_DESPESA) FROM DESPESA_PUBLICA NATURAL JOIN DETALHAMENTO_FUNCAO WHERE COD_FUNCAO IN(SELECT COD_FUNCAO FROM FUNCAO WHERE NOME_FUNCAO IN ('Educação', 'Saúde')) AND ANO_DESPESA = 2018 GROUP BY COD_ORGAO ORDER BY SUM(VALOR_DESPESA) DESC LIMIT 5)",conn)
df
```

```python
# 8 - ok
df = pd.read_sql_query("SELECT NOME_ORGAO, NOME_SUBFUNCAO, COUNT(*) AS QUANTIDADE_DESPESAS, SUM(VALOR_DESPESA) AS VALOR_DESPESAS FROM DESPESA_PUBLICA NATURAL JOIN DETALHAMENTO_FUNCAO NATURAL JOIN ORGAO NATURAL JOIN SUBFUNCAO GROUP BY COD_ORGAO, COD_SUBFUNCAO HAVING SUM(VALOR_DESPESA) > 10000 ORDER BY NOME_ORGAO, NOME_SUBFUNCAO",conn)
df
```

```python
# 2 - ok
df = pd.read_sql_query("SELECT NOME_FUNCAO, NOME_SUBFUNCAO, AVG(VALOR_DESPESA) AS MEDIA_DESPESAS FROM DESPESA_PUBLICA NATURAL JOIN DETALHAMENTO_FUNCAO NATURAL JOIN FUNCAO NATURAL JOIN SUBFUNCAO GROUP BY NOME_FUNCAO, NOME_SUBFUNCAO HAVING AVG(VALOR_DESPESA) > 1000",conn)
df
```

```python
# 3 - ok
df = pd.read_sql_query("SELECT NOME_PROGRAMA_GOVERNO, SUM(VALOR_DESPESA) FROM PROGRAMA_GOVERNO NATURAL JOIN DESPESA_PUBLICA WHERE ANO_DESPESA = 2018 GROUP BY COD_PROGRAMA_GOVERNO ORDER BY SUM(VALOR_DESPESA) DESC",conn)
df
```

```python
# 4 - ok
df = pd.read_sql_query("SELECT MES_DESPESA, ANO_DESPESA, MODALIDADE_DESPESA, NOME_PROGRAMA_GOVERNO, VALOR_DESPESA FROM DESPESA_PUBLICA NATURAL JOIN PROGRAMA_GOVERNO WHERE COD_ORGAO = (SELECT COD_ORGAO FROM ORGAO WHERE NOME_ORGAO = 'Ministério da Educação') AND MES_DESPESA <= 6 AND ANO_DESPESA = 2018 ORDER BY MES_DESPESA, ANO_DESPESA",conn)
df
```

```python
# 12 - ok
df = pd.read_sql_query("SELECT DISTINCT NOME_FUNCAO, VALOR_DESPESA FROM FUNCAO NATURAL JOIN DETALHAMENTO_FUNCAO NATURAL JOIN DESPESA_PUBLICA",conn)
df
```

```python
# 13 - ok
df = pd.read_sql_query("SELECT DISTINCT MODALIDADE_DESPESA, VALOR_DESPESA, NOME_ORGAO FROM (ORGAO NATURAL JOIN (SELECT * FROM DESPESA_PUBLICA WHERE MODALIDADE_DESPESA = 'Reserva de Contingência'));",conn)
df
```

```python
# 14 - ok
df = pd.read_sql_query("SELECT DISTINCT VALOR_DESPESA, NOME_PROGRAMA_GOVERNO FROM (DESPESA_PUBLICA NATURAL JOIN (SELECT * FROM PROGRAMA_GOVERNO WHERE NOME_PROGRAMA_GOVERNO = 'ENEM'))",conn)
df
```

```python
# 15 - ok
df = pd.read_sql_query("SELECT DISTINCT NOME_PROGRAMA_GOVERNO, QTD_ORGAOS_SUBORDINADOS, VALOR_DESPESA FROM (DESPESA_PUBLICA NATURAL JOIN (SELECT * FROM ORGAO WHERE NOME_ORGAO = 'Ministério da Educação') NATURAL JOIN PROGRAMA_GOVERNO)",conn)
df
```
